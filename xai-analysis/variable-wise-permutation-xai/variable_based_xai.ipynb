```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy
import os
import os.path 
import matplotlib.pyplot as plt
import matplotlib.colors
import seaborn as sns
import json
import random
```

## Results based on F file 

```python
permut = pd.read_csv('./f.csv')
group_scores = list(permut.groupby('groups')['Score'].sum())
group_names  = ['G1', 'G2', 'G3', 'G4', 'G5']

group_df = pd.DataFrame(columns = ['Group', 'Score'])
group_df['Group'] = group_names
group_df['Score'] = group_scores

fig, axs = plt.subplots(figsize = (8, 6))
ax = sns.barplot(x="Group", y="Score", data=group_df, ci=None, ax = axs)
plt.tight_layout()
```

```python
fig, axs = plt.subplots(figsize = (20, 8))
ax = sns.barplot(x="feature", y="Score", data=permut, hue= 'groups', ci=None, ax = axs)
plt.xticks(rotation=90, fontsize = 12)
plt.tight_layout()
```

### New

```python
df = pd.read_csv('./features.csv')
df['hss_score'] = abs(0.52 - df['HSS_Mean'])
df['pss_score'] = abs(0.55 - df['PSS_Mean'])
df
```

```python
hss_group_scores = list(df.groupby('group')['hss_score'].sum())
group_names  = ['G1', 'G2', 'G3', 'G4', 'G5']

hss_group_df = pd.DataFrame(columns = ['Group', 'Score'])
hss_group_df['Group'] = group_names
hss_group_df['Score'] = hss_group_scores
hss_group_df
```

```python
fig, axs = plt.subplots(figsize = (8, 6))
ax = sns.barplot(x="Group", y="Score", data=hss_group_df, ci=None, ax = axs)
plt.tight_layout()
```

```python
fig, axs = plt.subplots(figsize = (20, 8))
hss = sns.barplot(x="feature", y="hss_score", data=df, hue= 'group', ax = axs)
axs.set_ylim(0, 0.035)
plt.xticks([])
None
```
